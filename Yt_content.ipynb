```python
import pandas as pd
import requests
import re
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, classification_report

# Function to fetch the channel name from a YouTube video URL
def get_channel_name_from_url(youtube_url):
    response = requests.get(youtube_url)

    if response.status_code == 200:
        channel_name_pattern = re.compile(r'"ownerChannelName":"(.*?)"')
        matches = channel_name_pattern.search(response.text)

        if matches:
            channel_name = matches.group(1)
            return channel_name
        else:
            return "Channel name not found on the page."
    else:
        return f"Failed to fetch the YouTube page. Status code: {response.status_code}"


df = new_df

# Get the YouTube video URL from the user
url = input("Enter the URL: ")

# Get the channel name from the URL
channel_name = get_channel_name_from_url(url)
print(f"Channel Name: {channel_name}")

# Replace NaN values in the 'channel_type' column with 'Unknown'
df['channel_type'].fillna('Unknown', inplace=True)

# Encode categorical features
encoder = OrdinalEncoder()
data_encoded = encoder.fit_transform(df)

# Split the data into features (X) and target (y)
X = data_encoded[:, :-1]  # Exclude the last column ('channel_type')
y = data_encoded[:, -1]   # The last column ('channel_type') is the target

# Initialize and train the CategoricalNB model on the entire dataset
clf = CategoricalNB()
clf.fit(X, y)

# Predict the labels for the test set
y_pred = clf.predict(X)

# Evaluate the model's performance
accuracy = accuracy_score(y, y_pred)
classification_rep = classification_report(y, y_pred)

print(f"Accuracy: {accuracy:.2f}")

# Predict the category for the new Youtuber based on the provided channel name
new_youtuber = channel_name
predicted_category = None

for index, row in df.iterrows():
    if row['Youtuber'] == new_youtuber:
        predicted_category = row['channel_type']
        break

if predicted_category is None:
    predicted_category = 'Unknown'

print(f"Predicted Category: {predicted_category}")
```
